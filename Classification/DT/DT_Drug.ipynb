```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics
from io import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
```

```python
data = "../DataSets/drug200.csv"
df = pd.read_csv(data, delimiter=",")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.size
```

```python
x = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
```

```python
x
```

```python
lab_sex = preprocessing.LabelEncoder()
lab_sex.fit(['F', 'M'])
x[:,1] = lab_sex.transform(x[:,1])
```

```python
x
```

```python
lab_BP = preprocessing.LabelEncoder()
lab_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
x[:,2] = lab_BP.transform(x[:,2])
```

```python
x
```

```python
lab_chol = preprocessing.LabelEncoder()
lab_chol.fit([ 'NORMAL', 'HIGH'])
x[:,3] = lab_chol.transform(x[:,3])
```

```python
x
```

```python
y = df["Drug"]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 3 )
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
drugTree = DecisionTreeClassifier(criterion = "gini", max_depth = 4)
```

```python
drugTree.fit(x_train, y_train)
```

```python
y_predicted = drugTree.predict(x_test)
```

```python
y_predicted
y_test
```

```python
print(metrics.accuracy_score(y_test, y_predicted))
```

```python
dotData = StringIO()
filename = "drugtree.png"
featureName = df.columns[0:5]
output = tree.export_graphviz(drugTree, feature_names = featureName, out_file = dotData, class_names = np.unique(y_train), filled = True, special_characters = True, rotate =False)
graph = pydotplus.graph_from_dot_data(dotData.getvalue())
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img,interpolation='nearest')
```
