```python
#Dataframe manipulation library
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python
#Storing the movie information into a pandas dataframe
dataMovie = "/content/movie.csv"
dfMovie = pd.read_csv(dataMovie)

#Head is a function that gets the first N rows of a dataframe. N's default is 5.
dfMovie.head()
```

```python
#Storing the user information into a pandas dataframe
dataRating ="/content/ratings.csv"
dfRatings = pd.read_csv(dataRating)

#Head is a function that gets the first N rows of a dataframe. N's default is 5.
dfRatings.head()
```

```python
#Using regular expressions to find a year stored between parentheses
#We specify the parantheses so we don't conflict with movies that have years in their titles
dfMovie['year'] = dfMovie.title.str.extract('(\(\d\d\d\d\))',expand=False)
dfMovie.head()
```

```python
#Removing the parentheses
dfMovie['year'] = dfMovie.year.str.extract('(\d\d\d\d)',expand=False)
dfMovie.head()
```

```python
#Removing the years from the 'title' column
dfMovie['title'] = dfMovie['title'].str.replace(r"\s*\(\d+\)?", "", regex=True)
dfMovie.head()
```

```python
#Every genre is separated by a | so we simply have to call the split function on |
dfMovie['genres'] = dfMovie.genres.str.split('|')
dfMovie.head()
```

```python
#Copying the movie dataframe into a new one since we won't need to use the genre information in our first case.
moviesWithGenres_df = dfMovie.copy()

for index, row in dfMovie.iterrows():
    for genres in row['genres']:
        moviesWithGenres_df.at[index, genres] = 1
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()
```

```python
#Drop removes a specified row or column from a dataframe
dfRatings = dfRatings.drop('timestamp', axis = 1)
dfRatings.head()
```

```python
userInput = [
    {'title': 'Breakfast Club, The', 'rating':5},
    {'title':'Toy Story', 'rating':3.5},
    {'title':'Jumanji', 'rating':2},
    {'title':"Pulp Fiction", 'rating':5},
    {'title':'Akira', 'rating':4.5}
]

dfinputMovies = pd.DataFrame(userInput)
dfinputMovies
```

```python
#Filtering out the movies by title
inputId = dfMovie[dfMovie['title'].isin(dfinputMovies['title'].tolist())]
```

```python
#Then merging it so we can get the movieId. It's implicitly merging it by title.
dfinputMovies = pd.merge(inputId, dfinputMovies)
```

```python
#Dropping information we won't use from the input dataframe
dfinputMovies = dfinputMovies.drop('genres', axis = 1)
dfinputMovies = dfinputMovies.drop('year', axis = 1)

#Final input dataframe
#If a movie you added in above isn't here, then it might not be in the original
#dataframe or it might spelled differently, please check capitalisation.
dfinputMovies.head()
```

```python
#Filtering out the movies from the input
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(dfinputMovies['movieId'].tolist())]
userMovies.head()
```

```python
#Resetting the index to avoid future issues
userMovies = userMovies.reset_index(drop = True)
```

```python
#Dropping unnecessary issues due to save memory and to avoid issues
userGenreTable = userMovies.drop(['movieId', 'title', 'genres', 'year'], axis = 1)
userGenreTable.head()
```

```python
dfinputMovies['rating']
```

```python
#Dot produt to get weights
userProfile = userGenreTable.transpose().dot(dfinputMovies['rating'])
```

```python
#The user profile
userProfile.head()
```

```python
#Now let's get the genres of every movie in our original dataframe
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
```

```python
#drop the unnecessary information
genreTable = genreTable.drop(['movieId', 'title', 'genres', 'year'], axis = 1)
genreTable.head()
```

```python
genreTable.shape
```

```python
#Multiply the genres by the weights and then take the weighted average
dfRecommendation = ((genreTable * userProfile).sum(axis = 1)) / (userProfile.sum())
dfRecommendation.head()
```

```python
#Sort our recommendations in descending order
dfRecommendation = dfRecommendation.sort_values(ascending = False)
dfRecommendation.head()
```

```python
#The final recommendation table
dfMovie.loc[dfMovie['movieId'].isin(dfRecommendation.head(20).keys())]
```
